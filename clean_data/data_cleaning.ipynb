# Import the data

```python
df_purchaseBehaviour = pd.read_csv("https://raw.githubusercontent.com/aps0611/Quantium_Business_CASE/main/raw_data/QVI_purchase_behaviour.csv")
df_purchaseBehaviour.head(5)
```

```python
df_transactionData = pd.read_excel("https://raw.githubusercontent.com/aps0611/Quantium_Business_CASE/main/raw_data/QVI_transaction_data.xlsx")
df_transactionData.head(10)
```

## Modify the date column

```python
df_transactionData['DATE'] = pd.to_datetime(df_transactionData['DATE'], unit='D', origin='1900-01-01')
df_transactionData.head(5)
```

```python
# check the nan values in the dataset
df_transactionData.isna().sum()
```

```python
## extra spaces remove
df_transactionData['PROD_NAME'] = df_transactionData['PROD_NAME'].str.strip()
df_transactionData.head(5)
```

```python
df_transactionData['PROD_NAME'].unique()
```

```python
# lets create a new column with package size:

df_transactionData['PACK_SIZE'] = df_transactionData['PROD_NAME'].str[-4:]
df_transactionData.head(5)
```

```python
df_transactionData['PACK_SIZE'].unique()

## Salt is there in the array which we need to rectify
```

```python
df_transactionData['PACK_SIZE'].unique()
```

```python
df_transactionData['PROD_NAME'] = df_transactionData['PROD_NAME'].str.lower()
```

```python
df_transactionData['PROD_NAME'].unique()
```

```python
## remove multiple spaces


df_transactionData['PROD_NAME'] = df_transactionData['PROD_NAME'].str.replace(r'\s+', ' ', regex=True)
df_transactionData['PROD_NAME'].unique()
```

```python
elements_to_remove = ['175g', '170g', '150g', '300g', '330g', '210g', '210G', '270g',
       '220g', '125g', '110g', '134g', '150G', '175G', '380g', '180g',
       '165g', '135g', '250g', '200g', '160g', '190g', ' 90g', ' 70g']

# remove the package size from PROD_NAME Column
def remove_elements(row):
    for element in elements_to_remove:
        row = row.replace(element, '')
    return row

df_transactionData['PROD_NAME'] = df_transactionData['PROD_NAME'].apply(remove_elements)
```

```python
df_transactionData.head(5)
```

```python
brand_names = ['natural chip compny', 'ccs','smiths',
       'kettle','old el paso','grain waves','doritos',
       'ww', 'thins','burger rings','ncc','cheezels',
       'infzns','red rock deli','pringles',
       'natural chipco','grnwves','tyrrells',
       'cobs','woolworths','french fries','rrd',
       'tostitos', 'cheetos','snbts','sunbites','infuzions', 'smith',
        'twisties','dorito','natural chip co'
       ]

# Extract the brand names
def extract_brand_name(product_name):
    for brand in brand_names:
        if brand in product_name:
            return brand
    return 'Other'


df_transactionData['BRAND_NAME'] = df_transactionData['PROD_NAME'].apply(extract_brand_name)
df_transactionData.head(5)
```

```python
## There is a typo: Rectifying the same in brand names
brand_dict = {'natural chip compny':'ncc',
              'natural chipco':'ncc',
              'grain waves': 'sunbites',
              'grnwves':'sunbites',
              'snbts':'sunbites',
              'red rock deli':'rrd',
              'woolworths':'ww',
              'infzns':'infuzions',
              'smith':'smiths',
              'natural chip co':'ncc',
              'dorito':'doritos'
              }

df_transactionData['BRAND_NAME'].replace(brand_dict, inplace=True)
```

```python
df_transactionData.head(5)
```

```python
df_transactionData['PACK_SIZE']=df_transactionData['PACK_SIZE'].str.replace(r'[gG]','').astype(float)
```

```python
df_transactionData.head(10)
```

```python
# remove the package size from PROD_NAME Column
def remove_brands(row):
    for element in brand_names:
        row = row.replace(element, '')
    return row

df_transactionData['PROD_NAME'] = df_transactionData['PROD_NAME'].apply(remove_brands)
df_transactionData.head(5)
```

```python
df_transactionData['PROD_NAME'].unique()
```

```python
df_transactionData['PROD_NAME'] = df_transactionData['PROD_NAME'].str.strip()
df_transactionData.head()
```

```python
# check nan
df_transactionData.isna().sum()
```

```python
# check the unique brands
df_transactionData['BRAND_NAME'].unique()
```

```python
df_transactionData.loc[df_transactionData['BRAND_NAME'] == 'burger rings', 'PROD_NAME'] = 'burger rings'
df_transactionData[df_transactionData['BRAND_NAME'] == 'burger rings']
```

```python
# On visual inspection: Burger rings belong to the brand smiths snackfood co pvt ltd. Rectifying the same:
df_transactionData.loc[df_transactionData['BRAND_NAME'] == 'burger rings', 'BRAND_NAME'] = 'smiths'
```
